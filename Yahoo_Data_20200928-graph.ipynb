```python
import yfinance as yf
import pandas as pd
import numpy as np


import xlsxwriter
```

```python
ticker_data = yf.download('AAPL  GOOG', period = '1y')['Close']
df = pd.DataFrame(ticker_data)
df
```

```python
writer = pd.ExcelWriter('ticker_graph.xlsx', engine ='xlsxwriter')

df.to_excel(writer, sheet_name='FINAL VALUES', index=True)

workbook = writer.book
worksheet = writer.sheets['FINAL VALUES']

chart = workbook.add_chart({'type': 'line'})

ending = len(df)+2
chart.add_series({
    'name':       "='FINAL VALUES'!$B$1",
    'categories': "='FINAL VALUES'!$A$2:A$"+ str(ending),
    'values': "='FINAL VALUES'!$B$2:$B$"+ str(ending),
    'line': {'width': 1.75,},
#'marker': {'type': 'circle'},
#'data_labels': {'value': True},
})

chart.set_title ({'name': 'Line Plot Test'})
chart.set_x_axis({'name': 'date'})
chart.set_y_axis({'name': 'closing price'})

worksheet.insert_chart('D7', chart)

writer.save()
```
